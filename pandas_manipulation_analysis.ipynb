We use `&` instead of `and` when working with Pandas DataFrames because:

1. `&` is designed to perform element-wise operations on boolean arrays/series in Pandas
2. `and` is a Python logical operator that can only evaluate two boolean values, not entire arrays

If you tried to use `and` with Pandas boolean indexing, you would get an error because:
- `and` tries to evaluate the entire boolean series as a single True/False value
- `&` correctly compares each row's values individually
